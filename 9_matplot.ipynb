# Matplot Library

        Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002.

    One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

```python
### import matplot library
import matplotlib.pyplot as plt

# to show the result of the code or to give visualise the code in jupyter notebook perticularly we have to use plt.show to every code .
# so insted of doing that smply use %matplotlib inline

%matplotlib inline
```

```python
## import numpy
import numpy as np
```

```python
# some examples
# scatter plot

a=np.arange(0,10)
b=np.arange(11,21)
```

```python
plt.scatter(a,b,c='Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2-D')
plt.savefig('Testfig.png')
```

```python
#### plt plot
plt.plot(a,b)
```

```python
plt.plot(a,b*b)
```

```python
plt.plot(a,b,'r')
```

```python
plt.plot(a,b,'r--')
```

```python
plt.plot(a,b,'r*',linestyle='dashed')
# and many more format
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(a,b,'r')
plt.subplot(2,2,2)
plt.plot(a,b*b,'r--')
```

```python
x =np.arange(1,11)
y  = 3 * x + 5
plt.title('Matplot table')
plt.xlabel('X-axis')
plt.ylabel('y-Axis')
plt.plot(x,y,'r*--')
```

```python
np.pi
```

```python
x = np.arange(0,3*np.pi,0.5)
y =np.sin(x)
plt.title('Sine Wave form')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
```

### bar plot

```python
x = [1,2,3]
y = [4,5,6]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.bar(x,y)
```

## Histogram

```python
# Histogram will helf to give density
a = [22,22,24,25,26,89,75,45,56,29]
plt.hist(a)
plt.title('Histogram')
```

## Box plot using MatplotLib

```python
data = [np.random.normal(0 , std , 100)
        for std in range(1,4)]
print(data)
```

```python
# plot 
plt.boxplot(data)
```

### Pie Chart


```python
# Data to plot
labels = 'python','java', 'c','c++'
sizes = [100,100,100,100]
colors = ['g','y','r','b']
explode = (0.1,0,0,0) #explode 1st slice
#plot
plt.pie(sizes, explode = explode, labels = labels , colors = colors ,autopct= '%1.1f%%', shadow = True)

```
